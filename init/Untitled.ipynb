```python
from sklearn import datasets
iris = datasets.load_iris()

target = iris.target
# for i in range(3):
#     print('class : %s , nb exmplaire : %s ' % (i , len(target[target ==  i])))


data = iris.data 
type(data) , data.ndim, data.shape
print(iris.feature_names)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib as mpl 
import numpy as np

fig = plt.figure(figsize= (8,4))
fig.subplots_adjust(hspace= 0.4, wspace=0.4)
ax1 = plt.subplot(1,3,1)

clist = ['violet','yellow','blue']
colors = [clist[c] for c in iris.target]

ax1.scatter(data[:,0], data[:,1], c = colors)
plt.xlabel("longueur du sepal ")
plt.ylabel("largueur du sepal ")

ax2 = plt.subplot(1,3,2)

ax2.scatter(data[:,2], data[:,3],c = colors)
plt.xlabel('longueur du petal')
plt.ylabel('largueur du petal')

ax3 = plt.subplot(1,3,3)

ax3.scatter(data[:,0], data[:,2],c = colors)
plt.xlabel('largueur du petal')
plt.ylabel('largueur du petal')

for ind,s in enumerate(iris.target_names):
    plt.scatter([],[],label = s, color = clist[ind])
    
plt.legend(scatterpoints = 1, labelspacing=1, bbox_to_anchor=(1.8, .5), loc='center right', title ="Especes")
plt.plot()
```

```python
import seaborn as sns
import pandas as pd
sns.set()
df = pd.DataFrame(data,columns = iris['feature_names'])
df['target'] = target
df['label'] = df.apply(lambda x : iris['target_names'][int(x.target)], axis=1)
print(df.head())

sns.pairplot(df, hue='label', vars = iris['feature_names'], size =2 )
```

```python
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
clf.fit(data,target)
clf.get_params()
```

```python
result = clf.predict(data)
error = sum (result != target)
print('nb errors : ', error)
```
